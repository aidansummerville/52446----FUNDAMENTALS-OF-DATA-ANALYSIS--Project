## Terminolgy


### The Simple Boxplot

The boxplot displays numeric data within a dataset. Boxplots display variations on data through there quartiles. The simple boxplot displays data using five numeric values of minimum, first quartile, median, third quartile, maximum. Box plots give the user a view of the way in which data is distrubuted within a dataset

![Simple Box Plots](img/box1.gif)

####  Minimum

The minimum point represents the lowest number within the dataset. It is shown on the boxplot at the end of extended line from the bottom of the rectangle.


####  First Quartile

The first quartile extends from the bottom of the box to the median (the horizontal line in the middle of the box). The area inbetween the first quartile and the median represents 25% of the total plot points.

#### Median

The median represents the the middle point of the dataset. This point is represented by the the the horizontal line in the middle of the rectangle. This point is an average point within the dataset. 50% of the data lies above or below the this point.

#### Third Quartile

The third quartile extends from the top of the box to the median (the horizontal line in the middle of the box). The area inbetween the third quartile and the median represents 25% of the total plot points.

#### Maximum

The maximum point represents the highest number within the dataset. It is shown on the boxplot at the end of extended line from the top of the rectangle.

### Turkeys Boxplot

John Turkeys variant on the boxplot takes into account the fact that many datasets posses outliers that may skew the the dataset. Turkeys boxpot keeps the whiskers of the boxplot to 1.5 IQR ("Interquartile Range") The interquatile range is the distance between the "first quartile" and the "third quartile" 50% of the dataset centred around the mean. Anything outside of this is considered an oulier. If the point is within another 1.5 IQR (inbetween the "inner fence" and "outer fence")it is called a "suspected oulier" and is marked by a cear circle. Any point outside of this is called an "outlier"(outside the "outer fence") thease points are marked by black circles. Using this version of the boxplot allows the user to get a better understanding of the distrubution of the dataset that is not skewed by ouliers within the data.

![Outliers Box Plots](img/box2.gif)

#### Outliers 

Outliers are points of data 3 Interquartile Ranges outside of eithier the third or first quartile. There are represented by black filled circles. 

#### Suspected Outliers

Suspected outliers are 1.5 Interquartile range outside of eithier the third or first quartile. There are represented by clear  circles.

#### IQR  (Interquartile Range)

The Interquartile Range is The range between the first quartile and the third quartile. This range contains 50% of the total plots 25% below the median and 25% above the median. The Interquartile range is represented by the rectangle or box on the boxplot.

#### Inner Fence

The inner fence is the point at which points change between being part of the whisker to part if the suspected ouliers. This point is at 1.5 Interquartile range outside of eithier the third or first quartile. 

#### Outer Fence

The outer fence is the point at which points data changes from being represented as a suspected outliers to ouliers on the boxplot. This occurs at 3 IQR from eithier the third or first quartile.
