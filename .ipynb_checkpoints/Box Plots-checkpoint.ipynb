## History of Use

Box plots were developed by John Turkey above. The box plot can also be known as a box and whisker plot referring to the shape and makeup of the plot visible below. He introduced the box plot in his 1977 book Exploratory Data Analysis. While working at the famed Bell Labs Turkey coined the term bit and has many other accolades throughout his career including Turkeyâ€™s range test and Turkeys lambda distribution.

![Simple Box Plots](img/box3.png)

A box plot is a way of viewing the distribution of a dataset using the five number summary of minimum, first quartile, median, third quartile and maximum.
 - 	The median is the centre point of the box plot and the median of all the plot points. The median sits at 50%. This is the line at the centre of the box.
 - 	The First and Third Quartiles sit at 25% and 75% of the total distribution respectively. These points are represented by the edges of the box plot.
 - 	The Maximum and the Minimum Points are at the edge of the whiskers (lines that extend outward from the box plot). These points are located at 1.5 Interquartile Ranges or IQR(The range from the First quartile to the Third Quartile or the length of the box) from below the first quartile or above the third quartile(The edge of the box)
 - Any points outside of the maximum are minimum are displayed individually and are called outliers.

## Use Cases

I initially choose to use the Iris dataset for displaying the functions of the dataset. However while using this dataset it became clear that the type of data I was using was limiting the functionality of displaying a box plots capabilities so I proceeded to switch to the mart.csv file that I am using below. This dataset is made of sale attributes of cattle in the Headford mart in the previous 2 months and is the same dataset that I am using for my Programming for data analysis module.

```python

df = pd.read_csv("mart.csv", header=None) ## import dataset
```

```python
df.head() ## display first five lines of dataset
```

```python
df.columns = ["Lot", "Ring", "Weight", "No.Owners","Price", "Total", "Breed", "Sex","Price/Weight", "Price-Weight","Date"] ##  add column names
```

```python
df.head()  ## display first five lines of dataset
```

```python
df = df.loc[df.Breed.isin(['AAX', 'CHX', 'LMX'])] ## only taking thease three breeds into new dataset
```

The below box plot shows the distribution of the weight parameter within the dataset. From this simple box plot we can easily see:
- The median (The line through the centre of the box plot) is around the 510-520 mark
 - The second quartile (The edge of the box to the left) is around 470-480 mark
- The Third quartile (The edge of the box to the right) is around the 560 mark
- The maximum and minimum are at the 700 and 300 mark respectively
- The data is normally distributed while slightly skewed to the left
- There are many more outliers over the maximum rather than the minimum


```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x=df["Weight"])  ## displays weight of combined 3 breed in a boxplot with notch at the median
```

The box plots below show the distributions of weight for the three most sold breeds. While plotted side by side it is easier to see the relationships between the three breeds. All three breeds are similarly distributed with the AAX and CHX being more closely linked by their distributions than to the LMX breed. These plots also introduce the notches at the median. Notches are used to show similarities of different distributions means. If the different box plots notches do not overlap that means the means are too dissimilar to be called related. Notch sizes are directly related the size of the IQR or length of the box

```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Breed", y="Weight", hue="Breed",
              data=df, linewidth=2.5,notch=True) ## displays the weight of the 3 breeds individualy
```

The below box plot is similar to the above plot. It plots the three main breeds distributions also. However this box plot displays how the weight distributions changed over the course of a two month period. This kind of box plot is called a grouped box plot

```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Date", y="Weight", hue="Breed",
                  data=df, linewidth=2.5) ## displays the weights of the 3 breeds week on week in a boxplot
```

The below box plot shows again weight of breeds of cattle. In this box plot I have chosen what types of breeds and in what order to display myself using the order parameter.

```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Breed", y="Weight", data=df,
            order=["CHX", "LMX"]) ## displays chx and lmx breeds in boxplot format
```

The below box plot plots the all numeric variables within the box plot. The variables of No.Owners and Price over weight are so vastly minimal they are unable to be viewed. This is because the scale of the box plot is set to the variable of the highest value in this case total.

```python


plt.figure(figsize=(15,15))
ax = sns.boxplot(data=df, orient="h", palette="Set2") ## displays all integer parameters of the dataset in boxplot form
```

The below plot is something very commonly seen. It is a box plot with another kind of plot on top of it. In this case a swarm plot. Using plots in this way is very useful in displaying information that would not normally be seen while using a box plot on its own. Clusters of points either side of a mean with not much in-between would display similarly to a normally distributed box plot. On a normal box plot this would not be visible however in this kind of plot it would be seen immediately.

```python

plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Breed", y="Weight", hue="Breed",
              data=df, linewidth=2.5)
ax = sns.swarmplot(x="Breed", y="Weight", data=df, color=".25")   ## displays swarmplot and boxplots of weight of the breeds side by side
```

The below plot allows for a cat plot to be used in tandem with box plots to plot an additional variable of data at the same time. The below plot shows the weight distributions of the cattle across the number of owners. There is then the additional variable of breed plotted across three different graphs

```python

g = sns.catplot(x="No.Owners", y="Weight",
               hue="Breed", col="Breed",
                data=df, kind="box",
                height=10, aspect=.7);  ## displays weight of each breed broken down by number of owners
```

## Terminolgy


### The Simple Boxplot

The box plot displays numeric data within a dataset. Box plots display variations on data through there quartiles. The simple box plot displays data using five numeric values of minimum, first quartile, median, third quartile, maximum. Box plots give the user a view of the way in which data is distributed within a dataset

![Simple Box Plots](img/box1.gif)

####  Minimum

The minimum point represents the lowest number within the dataset. It is shown on the box plot at the end of extended line from the bottom of the rectangle.

####  First Quartile

The first quartile extends from the bottom of the box to the median (the horizontal line in the middle of the box). The area in-between the first quartile and the median represents 25% of the total plot points.

#### Median

The median represents the middle point of the dataset. This point is represented by the horizontal line in the middle of the rectangle. This point is an average point within the dataset. 50% of the data lies above or below this point.

#### Third Quartile

The third quartile extends from the top of the box to the median (the horizontal line in the middle of the box). The area in-between the third quartile and the median represents 25% of the total plot points.

#### Maximum

The maximum point represents the highest number within the dataset. It is shown on the box plot at the end of extended line from the top of the rectangle.

### Turkeys Boxplot

John Turkeys variant on the box plot takes into account the fact that many datasets posses outliers that may skew the  dataset. Turkeyâ€™s box pot keeps the whiskers of the box plot to 1.5 IQR ("Interquartile Range") The interquatile range is the distance between the "first quartile" and the "third quartile" 50% of the dataset centred around the mean. Anything outside of this is considered an outlier. If the point is within another 1.5 IQR (in-between the "inner fence" and "outer fence")it is called a "suspected outlier" and is marked by a clear circle. Any point outside of this is called an "outlier"(outside the "outer fence") these points are marked by black circles. Using this version of the box plot allows the user to get a better understanding of the distribution of the dataset that is not skewed by outliers within the data.

![Outliers Box Plots](img/box2.gif)

#### Outliers 

Outliers are points of data 3 Interquartile Ranges outside of either the third or first quartile. There are represented by black filled circles.

#### Suspected Outliers

Suspected outliers are 1.5 Interquartile range outside of either the third or first quartile. There are represented by clear circles.

#### IQR  (Interquartile Range)

The Interquartile Range is The range between the first quartile and the third quartile. This range contains 50% of the total plots 25% below the median and 25% above the median. The Interquartile range is represented by the rectangle or box on the box plot.

#### Inner Fence

The inner fence is the point at which points change between being part of the whisker to part if the suspected outliers. This point is at 1.5 Interquartile range outside of either the third or first quartile.

#### Outer Fence

The outer fence is the point at which points data changes from being represented as a suspected outliers to outliers on the box plot. This occurs at 3 IQR from either the third or first quartile.

## Alternatives

### Violin Plot

The Violin plot is a method of plotting numeric data similarly to the box plot.
- The violin plot although being less popular than the box plot displays more information in the same space.
- 	Violin plots work similarly to box plots but display kernal density data on either side.
-	Violin plots are also capable of plotting two variables against each other on the same plot either side of the centre line.
-	Bulges and clusters of data points would be more easily seen on a violin plot than on a box plot.
-	The box pot remains more popular possible because of its already wideness of use and understanding far more people than the violin plot.


```python
ax = sns.violinplot(x=df["Weight"]) ## violinplot depicting weight
```

Violin plot of the distribution of weights of the different breeds of cattle

```python
plt.figure(figsize=(15,15))
ax = sns.violinplot(x="Breed", y="Weight", hue="Breed",
              data=df, linewidth=2.5,notch=True) ## violin plot depecting the weight of the 3 most sold breeds within the dataset 
```

The below plot shows the distribution weight of the three breeds of cattle of a two month time period

```python
plt.figure(figsize=(15,15))
ax = sns.violinplot(x="Date", y="Weight", hue="Breed",
                  data=df, linewidth=2.5)
```

Below is a cat plot mixed with a violin plot

```python
g = sns.catplot(x="No.Owners", y="Weight",
               hue="Breed", col="Breed",
                data=df, kind="violin",
                height=10, aspect=.7);
```

### Stripplot

Strip plots can be used to show the categorical information of a numerical plot. They are often used in conjunction with box or violin plots to represent the individual spread of data points within the distributions. Very similar to swarm plots however they allow numeric data points to overlap on the plot.

Violin and strip plot below.

```python
plt.figure(figsize=(15,15))
ax = sns.violinplot(x="Date", y="Weight", data=df,
                    inner=None, color=".8")
ax = sns.stripplot(x="Date", y="Weight", data=df, jitter=True)
```

Box plot and strip lot below

```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Weight", y="Date", data=df, whis=np.inf)
ax = sns.stripplot(x="Weight", y="Date", data=df,
                   jitter=True, color=".3")
```

### Swarmplot

Swarm plots can be used to show the categorical information of a numerical plot. They are often used in conjunction with box or violin plots to represent the individual spread of data points within the distributions. Very similar to strip plots however they do not allow numeric data points to overlap on the plot.

```python
plt.figure(figsize=(15,15))
ax = sns.swarmplot(x="Breed", y="Weight", hue="Breed",
              data=df)
```

```python
plt.figure(figsize=(15,15))
ax = sns.boxplot(x="Weight", y="Date", data=df, whis=np.inf)
ax = sns.swarmplot(x="Weight", y="Date", data=df, color=".2")
```

```python
plt.figure(figsize=(15,15))
ax = sns.violinplot(x="Date", y="Weight", data=df, inner=None)
ax = sns.swarmplot(x="Date", y="Weight", data=df,
                    color="white", edgecolor="gray")
```

### Histogram

Histograms are another alternative to box plots to show the distributions of a dataset. They are useful to visualize a dataset and have the advantage of familiarity with most users. The below histogram plots the distribution of the weight variable. It also uses the kernel density line for further visualisation.

```python
x = df['Weight']
sns.distplot(x,rug = true);
```

The below histogram adds the rug function to display the points at which the points appear on the plot.

```python
x = df['Weight']
sns.distplot(x,rug = True);
```

```python

y=df['Price']
sns.distplot(y);
```

```python
y=df['Price']
sns.distplot(y,rug=True);
```

### Kernel Density Plot

The kernel density plot is another alternative to the box plot for viewing the distributions of a variable.

```python
ax = sns.kdeplot(x, shade=True, color="r")
```

The below plot shows two kernel density plots of different variables and their relationship to one another.

```python
g = sns.jointplot("Weight", "Price", data=df, kind="kde",
                 color="m", height=7)## kde jointplot of weight vs price
```
